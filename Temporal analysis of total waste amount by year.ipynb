## Categories waste using domain knowledge

```python
food_and_beverage = ['Food Wrappers/Containers','Takeout Containers (Plastic) ','Takeout Containers (Foam) ',\
                    'Bottle Caps (Plastic) ','Bottle Caps (Metal)','Bottle Caps (plastic and metal - 2017 onwards)',\
                     'Lids (Plastic)','Straws, Stirrers (Straws from 2017 onwards)','Forks, Knives, Spoons (Utensils from 2017 onwards)',\
                     'Beverage Bottles (Plastic) 2 liters or less (Plastic bottles from 2017 onwards)',\
                     'Glass Beverage Bottles','Beverage Cans','Coffee cups','Cups and Plates (Paper)',\
                     'Cups and Plates (Plastic) (Plastic cup from 2017 onwards)','Cups and Plates (Foam)',\
                     '6-Pack Holders'
                     ]


water_activity = ['Buoys/Floats (fishing buoy, pot, or trap from 2017 onwards)', 'Fishing Nets',\
                  'Rope','Fishing Line','Fishing Lures/Light Sticks','Other Plastic Bottles (oil, bleach etc) (Containers for 2017 onwards)',\
                  'Strapping Bands','55-Gallon Drums','Pallets','Bait Containers','Crab/Lobster/Fish Traps'       
                 ]


other = ['Bags (Plastic) (Plastic bags from 2017 onwards)','Other plastic bags','Bags (Paper) (Paper from 2017 onwards)',\
         'Other  Plastic/Foam Packaging (Other/Misc Packaging 2017 onwards)','Appliances (refrigerators, washers, etc.)',\
         'Balloons','Building Materials','Tobacco Packaging/Wrappers','Cigar Tips','Cigarette Lighters',\
         'Fireworks','Tires','Toys','Batteries','Clothing, Shoes','Foam (2017 onwards)','Foam Pieces','Plastic Pieces',\
         'Glass Pieces','Tiny Plastic or Foam (2017 onwards)','Other','Car/Car Parts','Light Bulbs','Plastic Sheeting',\
         'Crates','Pull Tabs','Shotgun Shells','Personal hygiene (2017 onwards)','Condoms','Diapers','Syringes','Tampons/Tampon Applicators'  
        ]
```

```python
data = pd.read_excel('/Users/kacy/Downloads/datajam.xlsx')

df2 = data
df2['food_and_beverage'] = df2[food_and_beverage].sum(axis = 1)
df2['water_activity'] = df2[water_activity].sum(axis = 1)
df2['other'] = df2[other].sum(axis = 1)

df2
```

```python
c2 = df2.groupby('Year').sum()
c2
```

```python
## locate each category
df_c2 = c2.iloc[:, -3:]
df_c2
```

## 2017-2019 change in each category

```python
plt.rcParams["figure.figsize"] = [10, 10]
a = df_c2.plot(kind="bar")
a.legend(loc=2, prop={'size': 8})
```
