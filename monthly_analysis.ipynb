```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
data = pd.read_excel("2017-2019 Vancouver Datajam .xlsx")
data = data.fillna(0)

data
```

```python
categories = data.columns[21:]
waste = data[categories]
date = data[['Year', 'Cleanup date']]
people = data['Adults (2017 onwards Participant Count)']

clean_data = pd.concat([date, waste, people], axis=1)

clean_data
```

```python
clean_data['Total Waste'] = clean_data[categories].sum(axis=1)

def format_time(time):
    try:
        month = str(time).split(' ')[0].split('-')[1]
    except:
        month = '01'
    return month

clean_data['Month'] = clean_data['Cleanup date'].apply(format_time)
group_data = clean_data.groupby(['Year', 'Month']).sum().reset_index()

group_data
```

```python
group_data_2017 = group_data[group_data['Year'] == 2017]
group_data_2018 = group_data[group_data['Year'] == 2018]
group_data_2019 = group_data[group_data['Year'] == 2019]

plt.figure(figsize=(15,10))
plt.plot('Month', 'Total Waste', data=group_data_2017, label='2017')
plt.plot('Month', 'Total Waste', data=group_data_2018, label='2018')
plt.plot('Month', 'Total Waste', data=group_data_2019, label='2019')
plt.xlabel('Month')
plt.ylabel('Number of Waste')
plt.title('Total waste comparison by month of each year')
plt.legend()
plt.savefig('total_waste_by_month_yearly.png')
```

```python
plt.figure(figsize=(15,10))
plt.plot('Month', 'Total Waste', data=group_data_2017, label='2017')
plt.plot('Month', 'Total Waste', data=group_data_2018, label='2018')
plt.plot('Month', 'Total Waste', data=group_data_2019, label='2019')
plt.yscale('log')
plt.xlabel('Month')
plt.ylabel('Number of Waste (log scale)')
plt.title('Total waste comparison by month of each year')
plt.legend()
plt.savefig('total_waste_by_month_yearly_log.png')
```

#### The amount of waste is the most around summer time
#### We have gathered more waste each year

### But is it because of the number of cleanup operations are different each year?

```python
operations_each_month = clean_data.groupby(['Year', 'Month']).size().reset_index()
operations_each_month = operations_each_month.rename(columns={0: 'Number of cleanup'})

operations_each_month
```

```python
operations_each_month_2017 = operations_each_month[operations_each_month['Year'] == 2017]
operations_each_month_2018 = operations_each_month[operations_each_month['Year'] == 2018]
operations_each_month_2019 = operations_each_month[operations_each_month['Year'] == 2019]

plt.figure(figsize=(15,10))
plt.plot('Month', 'Number of cleanup', data=operations_each_month_2017, label='2017')
plt.plot('Month', 'Number of cleanup', data=operations_each_month_2018, label='2018')
plt.plot('Month', 'Number of cleanup', data=operations_each_month_2019, label='2019')
plt.xlabel('Month')
plt.ylabel('Number of Cleanup')
plt.title('Number of Cleanup by month of each year')
plt.legend()
plt.savefig('cleanup_by_month_yearly.png')
```

#### 2019 have the most amount of waste and it has the most number of cleanup events

### But what about summer vs non-summer? Can we say summer is worse than non-summer? non-summer  doesn't have many cleanup events.

```python
average_data = group_data.merge(operations_each_month, on=['Year', 'Month'])
average_data['Average waste per operation'] = average_data['Total Waste'] / average_data['Number of cleanup']

average_data
```

```python
average_data_2017 = average_data[average_data['Year'] == 2017]
average_data_2018 = average_data[average_data['Year'] == 2018]
average_data_2019 = average_data[average_data['Year'] == 2019]

plt.figure(figsize=(15,10))
plt.plot('Month', 'Average waste per operation', data=average_data_2017, label='2017')
plt.plot('Month', 'Average waste per operation', data=average_data_2018, label='2018')
plt.plot('Month', 'Average waste per operation', data=average_data_2019, label='2019')
plt.xlabel('Month')
plt.ylabel('Average waste per cleanup')
plt.title('Average waste per cleanup by month of each year')
plt.legend()
plt.savefig('average_waste_per_cleanup_by_month_yearly.png')
```

#### there is no relationship of summer vs non-summer

### But each event has different number of participants

```python
people_2017 = group_data[group_data['Year'] == 2017]
people_2018 = group_data[group_data['Year'] == 2018]
people_2019 = group_data[group_data['Year'] == 2019]

plt.figure(figsize=(15,10))
plt.plot('Month', 'Adults (2017 onwards Participant Count)', data=people_2017, label='2017')
plt.plot('Month', 'Adults (2017 onwards Participant Count)', data=people_2018, label='2018')
plt.plot('Month', 'Adults (2017 onwards Participant Count)', data=people_2019, label='2019')
plt.xlabel('Month')
plt.ylabel('Number of participants')
plt.title('Number of participants by month of each year')
plt.legend()
```

```python
group_data['Average waste per person'] = group_data['Total Waste'] / group_data['Adults (2017 onwards Participant Count)']

average_people_2017 = group_data[group_data['Year'] == 2017]
average_people_2018 = group_data[group_data['Year'] == 2018]
average_people_2019 = group_data[group_data['Year'] == 2019]

plt.figure(figsize=(15,10))
plt.plot('Month', 'Average waste per person', data=average_people_2017, label='2017')
plt.plot('Month', 'Average waste per person', data=average_people_2018, label='2018')
plt.plot('Month', 'Average waste per person', data=average_people_2019, label='2019')
plt.xlabel('Month')
plt.ylabel('Average waste per person')
plt.title('Average waste per person by month of each year')
plt.legend()
```
