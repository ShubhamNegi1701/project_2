```python
# industry
hospitality = ['Food Wrappers/Containers','Bottle Caps (plastic and metal - 2017 onwards)','Lids (Plastic)',\
             'Straws, Stirrers (Straws from 2017 onwards)','Forks, Knives, Spoons (Utensils from 2017 onwards)',\
             'Beverage Bottles (Plastic) 2 liters or less (Plastic bottles from 2017 onwards)', \
             'Cups and Plates (Plastic) (Plastic cup from 2017 onwards)']
household = ['Personal hygiene (2017 onwards)','Clothing, Shoes','Toys','Tires',\
             'Other Plastic Bottles (oil, bleach etc) (Containers for 2017 onwards)', \
             'Bags (Plastic) (Plastic bags from 2017 onwards)']
leisure = ['Buoys/Floats (fishing buoy, pot, or trap from 2017 onwards)','Fishing Nets','Balloons',\
          '6-Pack Holders']
other = ['Other','Tiny Plastic or Foam (2017 onwards)','Foam (2017 onwards)','Unusual Items','Coffee cups',\
        'Other  Plastic/Foam Packaging (Other/Misc Packaging 2017 onwards)']
non_plastic = ['Bags (Paper) (Paper from 2017 onwards)', 'Beverage Cans','Glass Beverage Bottles','Cigarettes/Cigarette Filters','Rope']
```

```python
df.fillna(0, inplace = True)
df['hospitality'] = df[hospitality].sum(axis = 1)
df['household'] = df[household].sum(axis = 1)
df['leisure'] = df[leisure].sum(axis = 1)
df['other'] = df[other].sum(axis = 1)
df['non_plastic'] = df[non_plastic].sum(axis = 1)
```

```python
df2 = df[['Site Name','Kilograms','hospitality','household','leisure','other','non_plastic']].groupby(['Site Name']).sum()
df3 = df[['Site Name','Latitude','Longitude']].groupby(['Site Name']).mean()
```

```python
df2 = df2.merge(df3, left_on = 'Site Name',right_on = 'Site Name')
df2.reset_index(inplace = True)
```

```python
len(df2.index)
```

```python
df2.Kilograms = df2.Kilograms.astype(int)
df2.dtypes
```

```python
df2.dropna(subset=['Kilograms'], inplace = True)
```

```python
lat_lon = []
for i in range(len(df2.index)):
    lat_lon.append([df2['Latitude'][i],df2['Longitude'][i]])
```

```python
df2['lat_lon'] = lat_lon
plastic = ['hospitality','household','leisure','other']
df2['plastic'] = df2['hospitality'] + df2['household'] + df2['leisure'] + df2['other']
df2.head()
```

```python
subset = df2[['Site Name','lat_lon', 'plastic']]
tuples = subset.to_numpy()
tuples
```

```python
## from https://stackoverflow.com/questions/56842575/how-to-display-averages-instead-of-counts-on-folium-markerclusters
class MarkerWithProps(folium.Marker):
    _template = Template(u"""
        {% macro script(this, kwargs) %}
        var {{this.get_name()}} = L.marker(
            [{{this.location[0]}}, {{this.location[1]}}],
            {
                icon: new L.Icon.Default(),
                {%- if this.draggable %}
                draggable: true,
                autoPan: true,
                {%- endif %}
                {%- if this.props %}
                props : {{ this.props }} 
                {%- endif %}
                }
            )
            .addTo({{this._parent.get_name()}});
        {% endmacro %}
        """)
    def __init__(self, location, popup=None, tooltip=None, icon=None,
                 draggable=False, props = None ):
        super(MarkerWithProps, self).__init__(location=location,popup=popup,tooltip=tooltip,icon=icon,draggable=draggable)
        self.props = json.loads(json.dumps(props))    



plastic_map = folium.Map(title = 'Plastic Waste', location=[56.130, -106.35], zoom_start=3.5)

icon_create_function = '''
    function(cluster) {
        var markers = cluster.getAllChildMarkers();
        var sum = 0;
        for (var i = 0; i < markers.length; i++) {
            sum += markers[i].options.props.items;
        }

        return L.divIcon({
             html: '<b>' + sum + '</b>',
             className: 'marker-cluster marker-cluster-small',
             iconSize: new L.Icon.Default()
        });
    }
'''

marker_cluster = plugins.MarkerCluster(icon_create_function=icon_create_function)

for marker_item in tuples:
    marker = MarkerWithProps(
        location=marker_item[1],
        props = { 'items': marker_item[2]},
        popup=marker_item[0] + ':' + str(marker_item[2]),
        icon=folium.Icon(color='orange'),
        
    )
    marker.add_to(marker_cluster)

marker_cluster.add_to(plastic_map)    

#m.save(os.path.join('results', '1000_MarkerCluster0.html'))
# 
plastic_map.save("plastic_map.html")
```
